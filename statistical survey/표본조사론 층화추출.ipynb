```python
pd.crosstab(survey[17],survey[18],margins=True).style.background_gradient(cmap='summer_r')
```

```python
a=survey[(survey[18] == "부경대학교") & (survey[17] == 1)].sample(8, random_state=1)
b=survey[(survey[18] == "부경대학교") & (survey[17] == 2)].sample(13, random_state=1)
c=survey[(survey[18] == "부경대학교") & (survey[17] == 3)].sample(11, random_state=1)
d=survey[(survey[18] == "부경대학교") & (survey[17] == 4)].sample(13, random_state=1)

e=survey[(survey[18] == "부산대학교") & (survey[17] == 1)].sample(11, random_state=1)
f=survey[(survey[18] == "부산대학교") & (survey[17] == 2)].sample(15, random_state=1)
g=survey[(survey[18] == "부산대학교") & (survey[17] == 3)].sample(12, random_state=1)
h=survey[(survey[18] == "부산대학교") & (survey[17] == 4)].sample(15, random_state=1)
```

```python
survey=pd.concat([a,b,c,d,e,f,g,h])
```

```python
survey
```

```python
survey.to_csv(r"survey_sampling.csv", encoding='utf-8-sig', index=False)
```
